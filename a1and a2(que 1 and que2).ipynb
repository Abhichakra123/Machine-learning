```python
#a1(question number 1)
import pandas as pd
import numpy as np

# first of all ...we need to load the data
purchase_data = pd.read_excel("Lab Session Data.xlsx", sheet_name="Purchase data")

# having only relevant data that we need from both matrix a and matrix c
products = purchase_data[['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']].values
totpayment = purchase_data[['Payment (Rs)']].values

# for printing the data
print("Matrix A (Products):")
print(products)

print("\nMatrix C (totpayment):")
print(totpayment)

# dimensionality and samples
dimensionality = products.shape[1]
samplecount = products.shape[0]

# this is for rank
matrixrank = np.linalg.matrix_rank(products)

# Compute pseudo-inverse and solve for prices
pseudo_inverse = np.linalg.pinv(products)
product_prices = pseudo_inverse @ totpayment

print("\nVector Space Dimensionality:", dimensionality)
print("Total Vectors in Dataset:", samplecount)
print("Rank of the Product Matrix:", matrixrank)
print("\nEstimated Cost per Product (Candy, Mango, Milk):")
print(product_prices.flatten())
```
